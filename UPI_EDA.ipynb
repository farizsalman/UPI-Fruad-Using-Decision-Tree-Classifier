```python
# importing the necessary libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder,MinMaxScaler,OrdinalEncoder
import joblib
```

```python
# loading the dataset
UPI=pd.read_csv(r'C:\Users\salma\OneDrive\Desktop\UPI\upi_transactions.csv')
UPI
```

```python
# checking for existing null values in the data frame
UPI.isna().sum()
```

```python
UPI['timestamp'].dtype
```

```python
# changing the data type of column into datetime
UPI['timestamp'] = pd.to_datetime(UPI['timestamp'])
```

```python
# splitting the datetime into singular columns
UPI['date']=UPI['timestamp'].dt.date
UPI['time']=UPI['timestamp'].dt.time
UPI['month']=UPI['timestamp'].dt.month
UPI['year']=UPI['timestamp'].dt.year
UPI['day']=UPI['timestamp'].dt.day
UPI['hour']=UPI['timestamp'].dt.hour
UPI['minute']=UPI['timestamp'].dt.minute
UPI['second']=UPI['timestamp'].dt.second
```

```python
# drop the previous timestamp
UPI=UPI.drop('timestamp',axis=1)
```

```python
UPI['transaction_status'].value_counts()
```

```python
# encoding the string values into numbers
# OrdinalEncoder is used in cases where more number of value categories are presnt in the column
# LabelEncoder is used where less number of value categories are present in the column
```

```python
t_type=OrdinalEncoder(categories=[['P2M','P2P','Recharge','Bill Payment']])
UPI['transaction type']=t_type.fit_transform(UPI[['transaction type']])
```

```python
m_category=OrdinalEncoder(categories=[['Grocery','Food','Shopping','Fuel','Other','Utilities','Transport','Entertainment','Healthcare','Education']])
UPI['merchant_category']=m_category.fit_transform(UPI[['merchant_category']])
```

```python
t_status=LabelEncoder()
UPI['transaction_status']=t_status.fit_transform(UPI['transaction_status'])
```

```python
oe_state=OrdinalEncoder(categories=[['Maharashtra','Uttar Pradesh','Karnataka','Tamil Nadu','Delhi','Telangana','Gujarat','Andhra Pradesh','Rajasthan','West Bengal']])
UPI['sender_state']=oe_state.fit_transform(UPI[['sender_state']])
```

```python
send_bank=OrdinalEncoder(categories=[['SBI','HDFC','ICICI','IndusInd','Axis','PNB','Yes Bank','Kotak']])
UPI['sender_bank']=send_bank.fit_transform(UPI[['sender_bank']])
```

```python
receive_bank=OrdinalEncoder(categories=[['SBI','HDFC','ICICI','IndusInd','Axis','PNB','Yes Bank','Kotak']])
UPI['receiver_bank']=receive_bank.fit_transform(UPI[['receiver_bank']])
```

```python
oe_day=OrdinalEncoder(categories=[['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']])
UPI['day_of_week']=oe_day.fit_transform(UPI[['day_of_week']])
```

```python
send_age=OrdinalEncoder(categories=[['18-25','26-35','36-45','46-55','56+']])
UPI['sender_age_group']=send_age.fit_transform(UPI[['sender_age_group']])
```

```python
receive_age=OrdinalEncoder(categories=[['18-25','26-35','36-45','46-55','56+']])
UPI['receiver_age_group']=receive_age.fit_transform(UPI[['receiver_age_group']])
```

```python
network_oe=OrdinalEncoder(categories=[['3G','4G','5G','WiFi']])
UPI['network_type']=network_oe.fit_transform(UPI[['network_type']])
```

```python
device_oe=OrdinalEncoder(categories=[['Android','iOS','Web']])
UPI['device_type']=device_oe.fit_transform(UPI[['device_type']])
```

```python
# saving the encoders
joblib.dump(t_type,'t_type.joblib')
joblib.dump(m_category,'m_category.joblib')
joblib.dump(t_status,'t_status.joblib')
joblib.dump(oe_state,'oe_state.joblib')
joblib.dump(send_bank,'send_bank.joblib')
joblib.dump(receive_bank,'receive_bank.joblib')
joblib.dump(oe_day,'oe_day.joblib')
joblib.dump(send_age,'send_age.joblib')
joblib.dump(receive_age,'receive_age.joblib')
joblib.dump(network_oe,'network_oe.joblib')
joblib.dump(device_oe,'device_oe.joblib')
```

```python
# dropping the not important columns after encoding
```

```python
UPI=UPI.drop(['date','time'],axis=1)
```

```python
UPI=UPI.drop(['hour_of_day','hour','minute','second','is_weekend'],axis=1)
```

```python
# extracting only the numerical part of transaction id in order to convert to number
UPI['transaction id'] = UPI['transaction id'].str.extract(r'(\d+)').astype(int)
```

```python
# confirming all the columns are in numbers
UPI.dtypes
```

```python
# dataframe after preprocessing
UPI
```

```python
# saving the data frame locally
UPI.to_csv('UPI_preprocessed.csv',index=False)
```
